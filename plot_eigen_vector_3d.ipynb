```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def plot_3dvector(vectors,xlim,ylim,zlim,colors):
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111, projection='3d')
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.set_zlim(zlim)
    for i in range(len(vectors)):
        v = vectors[i]
        ax.quiver(0, 0, 0, v[0], v[1], v[2], color=colors[i])
        #ax.quiver(0, 0, 0, [v[0],0,0], [0,v[1],0],[0,0,v[2]], color='gray')
    
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
  
    plt.show()
```

```python
matrix =np.array([
                    [1,1,1],
                    [1,7,5],
                    [3,5,4]])
```

```python
eigen_values,eigen_vectors  = np.linalg.eig(matrix) 
```

```python
print(eigen_values)
```

```python
print(eigen_vectors[:,[0]])
print()
print(eigen_values[0])
print()
print(eigen_vectors[:,[0]]* eigen_values[0])
print()
```

```python
print(eigen_vectors[:,[1]])
print()
print(eigen_values[1])
print()
print(eigen_vectors[:,[1]]* eigen_values[1])
print()
```

```python
print(eigen_vectors[:,[2]])
print()
print(eigen_values[2])
print()
print(eigen_vectors[:,[2]]* eigen_values[2])
print()
```

```python
plot_3dvector(np.array([
                            eigen_vectors[:,[0]],
                            eigen_vectors[:,[0]] * eigen_values[0],                            
                            eigen_vectors[:,[1]],
                            eigen_vectors[:,[1]] * eigen_values[1],
                            eigen_vectors[:,[2]],
                            eigen_vectors[:,[2]] * eigen_values[2]
                        ]) , 
                             xlim=[-1,2],
                             ylim=[-1,2],
                             zlim=[-1,2],
                             colors=["b","g","b","g","b","g"])
```

```python
plot_3dvector(np.array([
                            eigen_vectors[:,[0]],                                                      
                            eigen_vectors[:,[1]],                          
                            eigen_vectors[:,[2]]
                          
                        ]) , 
                             xlim=[-1,2],
                             ylim=[-1,2],
                             zlim=[-1,2],
                             colors=["b","b","b"])
```

```python
plot_3dvector(np.array([
                            eigen_vectors[:,[0]] * eigen_values[0],
                            eigen_vectors[:,[1]] * eigen_values[1],                            
                            eigen_vectors[:,[2]] * eigen_values[2]
                        ]) , 
                             xlim=[-1,2],
                             ylim=[-1,2],
                             zlim=[-1,2],
                             colors=["g","g","g"])
```
